```python
import numpy as np
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
nltk.download('stopwords')
print(stopwords.words('english'))
```

Data Pre-processing

```python
news_dataset = pd.read_csv('/content/drive/MyDrive/dataset/train.csv')
```

```python
news_dataset.shape
```

```python
news_dataset.head()
```

```python
news_dataset.isnull().sum()
```

```python
news_dataset = news_dataset.fillna('')
```

```python
news_dataset['content'] = news_dataset['author'] + ' ' + news_dataset['title']

news_dataset['content'].head()
```

```python
port_stem = PorterStemmer()

def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
X = news_dataset['content'].values
Y = news_dataset['label'].values

print(X)
print(Y)
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
print(X)
```

Training data


```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y ,random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#accuracy of training data
X_train_prediction= model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy score of the training data :', training_accuracy) # for binary classification logistic regression is best model
```

```python
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score of the testing data :', testing_accuracy)
```
